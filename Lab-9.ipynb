```python
#importing settings to create engine
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # to get the password without showing the input
password = getpass.getpass()
```

```python
# creating connection with my database
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila' # database_location/database_name
engine = create_engine(connection_string)
```

```python
data = pd.read_sql_query('SELECT * FROM rentals_may', engine)
data.head()
```

```python
# checking the number of rentals in may
query =('''select customer_id, count(rental_id) as "Number of rentals in May" from rentals_may 
    group by customer_id 
    order by count(rental_id) desc;''')
data_may =pd.read_sql_query(query,engine)
data_may
```

```python
# checking the number of rentals in june
query =('''select customer_id, count(rental_id) as "Number of rentals in June" from rentals_June 
    group by customer_id 
    order by count(rental_id) desc;''')
data_june =pd.read_sql_query(query,engine)
data_june
```

```python
# merging both dataframes with an outer join so they are side by side
data_merged = pd.merge(data_may,data_june, how='outer')
data_merged
```

```python
# seen some Nans so checking how many
data_merged.isna().sum()
```

```python
# taking the nulls as in the customer didnt rent in that month so replacing with 0
data_merged = data_merged.fillna(0)
data_merged
```
